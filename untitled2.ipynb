```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer, GPT2Config
from torch.utils.data import Dataset, DataLoader
model_name = "gpt2"
config = GPT2Config.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name, config=config)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
```

```python
topics = ["Pivot Tables", "Data Visualization", "Formulas"]
queries = [
    "How do I create a pivot table in Excel?",
    "What are some ways to visualize data in Excel?",
    "How do I use the VLOOKUP formula?",
]
```

```python
class ExcelQueryDataset(Dataset):
    def __init__(self, topics, queries, tokenizer, max_length):
        self.topics = topics
        self.queries = queries
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.topics)

    def __getitem__(self, idx):
        topic = self.topics[idx]
        query = self.queries[idx]

        topic_query = f"Topic: {topic}\nQuery: {query}"
        input_ids = self.tokenizer.encode(topic_query, max_length=self.max_length, truncation=True)

        return {"input_ids": torch.tensor(input_ids)}
```

```python
max_length = 100
dataset = ExcelQueryDataset(topics, queries, tokenizer, max_length)
data_loader = DataLoader(dataset, batch_size=2, shuffle=True)

# Fine-tuning loop
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-5)

for epoch in range(5):  # Replace with desired number of epochs
    for batch in data_loader:
        optimizer.zero_grad()
        input_ids = batch["input_ids"]  # Assuming you have a GPU

        outputs = model(input_ids, labels=input_ids)
        loss = outputs.loss
        loss.backward()
        optimizer.step()

# Save fine-tuned model
model.save_pretrained("excel_query_model")
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the fine-tuned model and tokenizer
model_name = "excel_query_model"
model = GPT2LMHeadModel.from_pretrained("excel_query_model")
tokenizer = GPT2Tokenizer.from_pretrained("a")

# Input a prompt or query
prompt = "Topic: Formulas\nQuery: How do I use the SUM function in Excel?"

# Tokenize and generate response
input_ids = tokenizer.encode(prompt, return_tensors="pt")
with torch.no_grad():
    output = model.generate(input_ids, max_length=100, num_return_sequences=1)

response = tokenizer.decode(output[0], skip_special_tokens=True)

print("Model Response:", response)
```
